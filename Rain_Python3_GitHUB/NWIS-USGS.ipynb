## Arange NWIS streamflow to regular space 

```python
import pandas as pd
from datetime import datetime

#08068720
#08068740
#08068780
#08068800
#08068900
#08069000

fileIn = "C:/temp/Flood_Forecasting/NWIS-2012/08069000.csv"
fileOut = "C:/temp/Flood_Forecasting/NWIS-2012/08069000_AR_USTime.csv"

# date range taken from USGS download script
start = '2001-01-01 00:00:00'
end = '2019-07-25 00:00:00'

# this is only a reference
idx = pd.date_range(datetime(2001, 1, 1, hour=0, minute=0), periods=617023, freq='15min', tz='US/Central') 

data = pd.read_csv(fileIn)
data['datetime'] = pd.to_datetime(data['datetime'], utc=True)
data = data.set_index('datetime')
data.index = pd.DatetimeIndex(data.index)
data = data.reindex(idx)
data.to_csv(fileOut)
data
```
